```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
years_df = pd.read_csv("https://raw.githubusercontent.com/avgupta456/statbotics-csvs/main/v2/years.csv")
team_years_df = pd.read_csv("https://raw.githubusercontent.com/avgupta456/statbotics-csvs/main/v2/team_years.csv")
```

```python
# for year in range(2002, 2023):
for year in range(2022, 2023):
    if year == 2021:
        continue
        
    df = team_years_df[team_years_df.year == year]
    SCORE_MEAN = years_df[years_df.year == year].score_mean.iloc[0]
    SCORE_SD = years_df[years_df.year == year].score_sd.iloc[0]

    unitless_epa = 1500 + 250 * (df.epa_end - SCORE_MEAN / 3) / SCORE_SD
    norm_epa = df.norm_epa_end

    fig, ax = plt.subplots()
    ax.scatter(norm_epa, unitless_epa, s=1, label="Unitless EPA")
    ax.scatter(norm_epa, norm_epa, s=1, label="Year Normalized EPA")
    ax.legend()
    fig.show()
```

### Result

The x-axis is Year Normalized EPA, available at the end of the year. The blue line shows Unitless EPA, which is a simple linear transformation applied to EPA and depends only on the Week 1 mean and standard deviation. This measurement could be available even before the season ends. The difference between the blue and orange line shows the change in unitless EPA to normalized EPA. Unfortunately, it is too large to justify using unitless EPA on the website, as the noticeable shift at the end of the year would be confusing to users.
